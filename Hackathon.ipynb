```python
#Fix Data Types 
df1["date"] = pd.to_datetime(df1["date"])
df2["date"] = pd.to_datetime(df2["date"])

df1["quantity_sold_kilo"] = df1["quantity_sold_kilo"].astype(float)
df1["unit_selling_price_rmb_kg"] = df1["unit_selling_price_rmb_kg"].astype(float)
df2["wholesale_price_rmb_kg"] = df2["wholesale_price_rmb_kg"].astype(float)
```

```python
df1.head()
```

```python
sales_df = df1[df1["sale_or_return"] == "sale"].copy()
```

```python
#Merge All Datasets

final_df = (
    sales_df
    .merge(df, on="item_code", how="left")
    .merge(df2, on=["item_code", "date"], how="left")
)
```

```python
final_df.isna().sum()
```

```python
# Revenue
final_df["revenue"] = (
    final_df["quantity_sold_kilo"] *
    final_df["unit_selling_price_rmb_kg"]
)
```

```python
# Cost
final_df["cost"] = (
    final_df["quantity_sold_kilo"] *
    final_df["wholesale_price_rmb_kg"]
)
```

```python
# Gross margin
final_df["gross_profit"] = final_df["revenue"] - final_df["cost"]
```
